```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=1.00e16
mu=1500
q=1.6e-19
```

```python
E=np.linspace(0,100,100)
```

```python
J_d=q*n*mu*E
```

```python
plt.plot(E,J_d)
plt.xlabel("Electric Field(V/m)")
plt.ylabel("Drift Current Density(A/M^2)")
plt.title("Drift Current Densaity vs Electric Field")
plt.grid(True)
plt.show()
```
