```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 상수 정의
q = 1.6e-19  # 전자의 전하량 (C)
n = 1.0e16  # 캐리어 농도 (cm^-3)
E = np.linspace(0, 100, 100)  # 전기장 (V/m)
```

```python
# 이동도 값들 (cm^2/Vs)
mobility_values = [500, 1000, 1500]
```

```python
# 그래프 그리기
plt.figure(figsize=(8, 6))
for mu in mobility_values:
    J_d = q * n * mu * E  # 드리프트 전류 밀도 (A/m^2)
    plt.plot(E, J_d, label="이동도 = " + str(mu) + " cm²/Vs")

plt.xlabel("전기장 (V/m)")
plt.ylabel("드리프트 전류 밀도 (A/m²)")
plt.title("이동도에 따른 드리프트 전류 밀도 변화")
plt.grid(True)
plt.legend()
plt.show()
```
