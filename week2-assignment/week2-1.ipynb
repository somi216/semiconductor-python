```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kB=8.62*10**(-5)
```

```python
def FD_dist(E, E_f,T):
    f=1/(1+np.exp((E-E_f)/(kB*T)))
    return f
```

```python
Es=np.linspace(0,3,1000)
```

```python
plt.plot(Es,FD_dist(Es,1,100), label='T=100')
plt.plot(Es,FD_dist(Es,1,300), label='T=300')
plt.plot(Es,FD_dist(Es,1,1000), label='T=1000')
plt.xlabel(r'E(ev)')
plt.ylabel('probability')
plt.title('Fermi-Dirac Distribution')
plt.legend()
plt.show()
```

```python
plt.plot(Es,FD_dist(Es,1,150), label='T=150')
plt.plot(Es,FD_dist(Es,1,200), label='T=200')
plt.plot(Es,FD_dist(Es,1,1300), label='T=1300')
plt.xlabel(r'E(ev)')
plt.ylabel('probability')
plt.title('Fermi-Dirac Distribution')
plt.legend()
plt.show()
```
