```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
k_B = 8.617e-5  # Boltzmann constant in eV/K
E_g = 1.1  # Band gap energy in eV
N_d = 1.00e17  # Donor concentration (cm^-3)
n_i = 1.50e10  # Intrinsic carrier concentration (cm^-3)
```

```python
# Temperature range (K)
T = np.linspace(100, 500, 100)
```

```python
# Calculate carrier density and Fermi level
n0 = N_d
p0 = (n_i**2) / n0
n = n0 - p0
E_f = k_B * T * np.log(n / n_i)
```

```python
# Plot Fermi level vs Temperature
plt.plot(T, E_f, label="Fermi Level")
plt.xlabel("Temperature (K)")
plt.ylabel("Fermi Level (eV)")
plt.title("Fermi Level vs Temperature")
plt.grid(True)
plt.legend()
plt.show()
```
